### Perceptron vs. Adaline vs. MultiLayerPerceptron (logistic sigmoid)

```python
#!pip install mlxtend
```

```python
# Importing python ml extension library and some more
from mlxtend.plotting import plot_decision_regions
from mlxtend.classifier import Perceptron, Adaline, MultiLayerPerceptron
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
import matplotlib.gridspec as gridspec
import itertools
```

```python
# Generating a grid and some nice moons
gs = gridspec.GridSpec(2, 2)
X, y = make_moons(n_samples=100, random_state=123)

# Initializing the different algorithms
ppn = Perceptron(epochs=50, eta=0.05, random_seed=0)
ada = Adaline(epochs=50, eta=0.05, random_seed=0)
mlp = MultiLayerPerceptron(n_classes=len(np.unique(y)),
                           hidden_layers=[150],
                           l2=0.0,
                           l1=0.0,
                           epochs=500,
                           eta=0.01,
                           momentum=0.0,
                           decrease_const=0.0,
                           minibatches=1,
                           random_seed=42)

# Fitting and plotting the algorithms
fig = plt.figure(figsize=(10,8))
for clf, lab, grd in zip([ppn, ppn, mlp],
                         ['Perceptron', 'Adaline', 'MLP (logistic sigmoid)'],
                         itertools.product([0, 1], repeat=2)):
    clf.fit(X, y)
    ax = plt.subplot(gs[grd[0], grd[1]])
    fig = plot_decision_regions(X=X, y=y, clf=clf, legend=2)
    plt.title(lab)

plt.show()
```
